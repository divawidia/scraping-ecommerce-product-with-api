```python
import requests
import pandas as pd
```

```python
store_id = "4368886"
```

```python
url = "https://gql.tokopedia.com/graphql/ShopProducts"
page = 1
product_list = []

while True:
    payload = [{
        'operationName': "ShopProducts",
        'variables': {
            "etalaseId": "etalase", 
            "page": page, 
            "perPage": 200,
            "sid":f"{store_id}", 
            "sort": 11,
            "user_cityId": "176",
            "user_districtId": "2274",
            "user_lat": "",
            "user_long": ""
        },
        'query': 'query ShopProducts($sid: String!, $page: Int, $perPage: Int, $keyword: String, $etalaseId: String, $sort: Int, $user_districtId: String, $user_cityId: String, $user_lat: String, $user_long: String) {\n  GetShopProduct(shopID: $sid, filter: {page: $page, perPage: $perPage, fkeyword: $keyword, fmenu: $etalaseId, sort: $sort, user_districtId: $user_districtId, user_cityId: $user_cityId, user_lat: $user_lat, user_long: $user_long}) {\n    status\n    errors\n    links {\n      prev\n      next\n      __typename\n    }\n    data {\n      name\n      product_url\n      product_id\n      price {\n        text_idr\n        __typename\n      }\n      primary_image {\n        original\n        thumbnail\n        resize300\n        __typename\n      }\n      flags {\n        isSold\n        isPreorder\n        isWholesale\n        isWishlist\n        __typename\n      }\n      campaign {\n        discounted_percentage\n        original_price_fmt\n        start_date\n        end_date\n        __typename\n      }\n      label {\n        color_hex\n        content\n        __typename\n      }\n      label_groups {\n        position\n        title\n        type\n        url\n        __typename\n      }\n      badge {\n        title\n        image_url\n        __typename\n      }\n      stats {\n        reviewCount\n        rating\n        __typename\n      }\n      category {\n        id\n        __typename\n      }\n      __typename\n    }\n    __typename\n  }\n}\n'
    }]

    response = requests.post(url, json=payload)

    data = response.json()

    for product in data[0]['data']['GetShopProduct']['data']:
        product_list.append(product)
    
    if data[0]['data']['GetShopProduct']['links']['next'] == '':
        break

    page += 1
```

```python
df = pd.json_normalize(product_list)
df
```

```python
url_review = "https://gql.tokopedia.com/graphql/productReviewList"
result_review = []

for i in range(0, len(product_list)):
    product = product_list[i]
    page_review = 1

    while True:
        payload_review = [{
            'operationName': "productReviewList",
            'variables': {
                "filterBy": "", 
                "limit": 500, 
                "page": page_review,
                "productID":f"{product_list[i]['product_id']}", 
                "sortBy": 'create_time desc'
            },
            'query': 'query productReviewList($productID: String!, $page: Int!, $limit: Int!, $sortBy: String, $filterBy: String) {\n  productrevGetProductReviewList(productID: $productID, page: $page, limit: $limit, sortBy: $sortBy, filterBy: $filterBy) {\n    productID\n    list {\n      id: feedbackID\n      variantName\n      message\n      productRating\n      reviewCreateTime\n      reviewCreateTimestamp\n      isReportable\n      isAnonymous\n      imageAttachments {\n        attachmentID\n        imageThumbnailUrl\n        imageUrl\n        __typename\n      }\n      videoAttachments {\n        attachmentID\n        videoUrl\n        __typename\n      }\n      reviewResponse {\n        message\n        createTime\n        __typename\n      }\n      user {\n        userID\n        fullName\n        image\n        url\n        __typename\n      }\n      likeDislike {\n        totalLike\n        likeStatus\n        __typename\n      }\n      stats {\n        key\n        formatted\n        count\n        __typename\n      }\n      badRatingReasonFmt\n      __typename\n    }\n    shop {\n      shopID\n      name\n      url\n      image\n      __typename\n    }\n    hasNext\n    totalReviews\n    __typename\n  }\n}\n'
        }]

        response_review = requests.post(url_review, json=payload_review)

        data_review = response_review.json()

        for review in data_review[0]['data']['productrevGetProductReviewList']['list']:
            review.update({"product":product})
            result_review.append(review)

        print("Scraped product no. "+str(i)+" id: " + product_list[i]['product_id'] + " review page:"+str(page_review))
        
        if data_review[0]['data']['productrevGetProductReviewList']['hasNext'] == False:
            break

        page_review += 1
    print("Scraping product no. "+str(i)+" id: " + product_list[i]['product_id'] + " review done")

```
