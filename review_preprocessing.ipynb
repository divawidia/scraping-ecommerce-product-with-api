```python
import cleantext
import demoji
```

```python
def bersihkan_review(text):
    text = cleantext.clean(
        text, 
        extra_spaces=True, 
        numbers=True, 
        lowercase=True,
        reg='[^\w\s]',
        reg_replace=' '
    )
    text = re.sub('\n', ' ', text)
    text = re.sub('&amp;', 'dan', text)
    remove_emoji = demoji.findall(text)

    for item in remove_emoji.keys():
        text = text.replace(item, '')

    return text
```

```python
df_review['review_bersih'] = df_review['review'].apply(lambda review: bersihkan_review(review))
df_review
```

```python
# normalization
import re
import string
import csv

reader = csv.reader(open('slang_word.csv', 'r'), delimiter=';')
d = {}
for row in reader:
    k,v= row
    d[k] = v
    #print d[k]
pat = re.compile(r"\b(%s)\b" % "|".join(d))
for i in range(len(df_review)):
    text = df_review['review_bersih'].iloc[i]
    text = pat.sub(lambda m: d.get(m.group()), text)
    #print text
    df_review['review_bersih'].iloc[i]=text
df_review
```

```python
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
 
factory = StopWordRemoverFactory()
stopword = factory.create_stop_word_remover()
 
```
