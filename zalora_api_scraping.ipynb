```python
url_review = 'https://www.zalora.co.id/productreviews/reviews/'
review_list = []

for index in range(0, len(product_list)):
    query_params_review = {
        'sku':f'{product_list[index]["meta"]["sku"]}',
        'token':'',
        'per_page':'40'
    }

    headers_review = {
        'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/107.0.0.0 Safari/537.36 Edg/107.0.1418.42',
    }

    response_review = requests.request("GET", url_review, headers=headers_review, params=query_params_review).json()
    product = product_list[index]
    try:
        for review in response_review['reviews']:
            review.update({"product":product})
            review_list.append(review)

        try:
            while response_review['cursor']['has_next']:
                query_params_review = {
                    'sku':f'{product_list[index]["meta"]["sku"]}',
                    'token':f'{response_review["cursor"]["next"]}',
                    'per_page':'40'
                }

                response_review = requests.request("GET", url_review, headers=headers_review, params=query_params_review).json()
                
                for review in response_review['reviews']:
                    review.update({"product":product})
                    review_list.append(review)
        except KeyError as e:
            print("Scraping product no. "+str(index)+" sku: " + product_list[index]["meta"]["sku"] + " review done")
    except KeyError as e:
        print("Product no. "+str(index)+" sku :" + product_list[index]["meta"]["sku"] + " has no review")
```
