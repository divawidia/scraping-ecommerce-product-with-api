```python
data_bukalapak = collection_bukalapak.find()

df_bukalapak = pd.json_normalize(data_bukalapak)
df_bukalapak = pd.DataFrame(df_bukalapak[['review.content', 'created_at', 'product.name', 'product.specs.merek']])
df_bukalapak['created_at'] = pd.to_datetime(df_bukalapak['created_at'])
df_bukalapak.rename(columns = {'review.content':'review', 'created_at':'review_date', 'product.name':'product_name', 'product.specs.merek':'product_brand'}, inplace = True)
df_bukalapak
```

```python
data_blili = collection_blibli.find()

df_blibli = pd.json_normalize(data_blili)
df_blibli = pd.DataFrame(df_blibli[['content', 'createdDate', 'product.name', 'product.brand']])
df_blibli['createdDate'] = pd.to_datetime(df_blibli['createdDate'], unit='ms')
df_blibli.rename(columns = {'content':'review', 'createdDate':'review_date', 'product.name':'product_name', 'product.brand':'product_brand'}, inplace = True)
df_blibli 
```

```python
data_zalora = collection_zalora.find()

df_zalora = pd.json_normalize(data_zalora)
df_zalora = df_zalora.to_dict('records')
df_zalora = pd.json_normalize(df_zalora, 'comments', ['product.meta.name', 'product.meta.brand'])
df_zalora = pd.DataFrame(df_zalora[['text', 'created_at', 'product.meta.name', 'product.meta.brand']])
df_zalora.rename(columns = {'text':'review', 'created_at':'review_date','product.meta.name':'product_name', 'product.meta.brand':'product_brand'}, inplace = True)
df_zalora
```

```python
data_tokopedia = collection_tokopedia.find()

df_tokopedia = pd.json_normalize(data_tokopedia)
df_tokopedia = pd.DataFrame(df_tokopedia[['message', 'reviewCreateTime', 'product.name']])
df_tokopedia['reviewCreateTime'] = pd.to_datetime(df_tokopedia['reviewCreateTime'], unit='s')
df_tokopedia.rename(columns = {'message':'review', 'reviewCreateTime':'review_date','product.name':'product_name'}, inplace = True)
df_tokopedia
```

```python
data_jdid = collection_jdid.find()

df_jdid = pd.json_normalize(data_jdid, 'commentChildrens', [['product','skuName']])
df_jdid['commentDate'] = pd.to_datetime(df_jdid['commentDate'], unit='ms')
df_jdid = pd.DataFrame(df_jdid[['content', 'commentDate', 'product.skuName']])
df_jdid.rename(columns = {'content':'review', 'commentDate':'review_date','product.skuName':'product_name'}, inplace = True)
df_jdid
```

```python
data_shopee = collection_shopee.find()

df_shopee = pd.json_normalize(data_shopee, 'product_items', ['comment', 'ctime'])
df_shopee = pd.DataFrame(df_shopee[['comment', 'ctime', 'name']])
df_shopee['ctime'] = pd.to_datetime(df_shopee['ctime'], unit='s')
df_shopee.rename(columns = {'comment':'review', 'ctime':'review_date','name':'product_name'}, inplace = True)
df_shopee
```

```python
import cleantext
import demoji
```

```python
def bersihkan_review(text):
    text = cleantext.clean(
        text, 
        extra_spaces=True, 
        numbers=True, 
        lowercase=True,
        reg='[^\w\s]',
        reg_replace=' '
    )
    text = re.sub('\n', ' ', text)
    text = re.sub('&amp;', 'dan', text)
    remove_emoji = demoji.findall(text)

    for item in remove_emoji.keys():
        text = text.replace(item, '')

    return text
```

```python
df_review['review_bersih'] = df_review['review'].apply(lambda review: bersihkan_review(review))
df_review
```

```python
# normalization
import re
import string
import csv

reader = csv.reader(open('slang_word.csv', 'r'))
d = {}
for row in reader:
    k,v= row
    d[k] = v
    #print d[k]
pat = re.compile(r"\b(%s)\b" % "|".join(d))
for i in range(len(df_review)):
    text = df_review['review_bersih'].iloc[i]
    text = pat.sub(lambda m: d.get(m.group()), text)
    #print text
    df_review['review_bersih'].iloc[i]=text
df_review
```
