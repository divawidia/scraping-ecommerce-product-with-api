```python
import cleantext
import demoji
```

```python
def bersihkan_review(text):
    text = cleantext.clean(text, extra_spaces=True, numbers=True, punct=True, lowercase=True)
    remove_emoji = demoji.findall(text)

    for item in remove_emoji.keys():
        text = text.replace(item, '')

    return text
```

```python
df_content['review_bersih'] = df_content['content'].apply(lambda review: bersihkan_review(review))
df_content
```

```python
# normalization
import re
import string
import csv

reader = csv.reader(open('normalisasi.csv', 'r'))
d = {}
for row in reader:
    k,v= row
    d[cleantext.clean(k, lowercase=True)] = cleantext.clean(v, lowercase=True)
    #print d[k]
pat = re.compile(r"\b(%s)\b" % "|".join(d))
for i in range(len(df_content)):
    text = df_content['review_bersih'].iloc[i]
    text = pat.sub(lambda m: d.get(m.group()), text)
    #print text
    df_content['review_bersih'].iloc[i]=text
df_content
```

```python
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
 
factory = StopWordRemoverFactory()
stopword = factory.create_stop_word_remover()
 
```

```python
review = []
for index, row in df_content.iterrows():
    review.append(stopword.remove(row["review_bersih"]))
    
df_content["hasil_stopword"] = review
df_content
```

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
factory = StemmerFactory()
stemmer = factory.create_stemmer()
```

```python
review = []
for index, row in df_content.iterrows():
    review.append(stemmer.stem(row["hasil_stopword"]))
    
df_content["hasil_stemming"] = review
df_content
```
