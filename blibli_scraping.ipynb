```python
url_review = "https://www.blibli.com/backend/product-review/public-reviews"

result_review = []

for i in range(0, len(product_list)):
    query_params_review = {
        "page":"1",
        "itemPerPage":"30",
        "productSku":f"{product_list[i]['sku']}"
    }
    headers_review = {
        'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/107.0.0.0 Safari/537.36 Edg/107.0.1418.42',
    }
    response_review = requests.request("GET", url=url_review, headers=headers_review, params=query_params_review).json()

    product = product_list[i]

    try:
        if response_review["paging"]["total_page"] > 1:
            for page in range(1, response_review["paging"]["total_page"]+1):
                querystring_review2 = {"page":f"{page}","itemPerPage":"30","productSku":f"{product_list[i]['sku']}"}

                response_review2 = requests.request("GET", url=url_review, headers=headers_review, params=querystring_review2)
                
                data_review2 = response_review2.json()

                for a in data_review2["data"]:
                    a.update({"product":product})
                    result_review.append(a)
        else:
            for a in response_review["data"]:
                a.update({"product":product})
                result_review.append(a)

        print("Scraping product no. "+str(i)+" sku: " + product_list[i]['sku'] + " review done")
    except Exception as e:
        print(response_review)
        print(e)

```
